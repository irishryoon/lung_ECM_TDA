# Analysis on one example image

```julia
# select image
LTX = "001"
Da = "232"
idx = 1
filename = "LTX" * LTX * "_Da" * Da * "_idx" * string(idx)

low = 0.00063768
high = 0.4025769
min_sample = 100
max_sample = 5000
ECM_directory = "data/4000x4000_combined/subregion_ECM/"
sample_CSV_directory = "image_resolution_sensitivity/";
```

```julia
factors = vcat([1], collect(4:4:128));
```

```julia
# for each factor, create downsampled images & compute PD
PD0_ECM = Dict()
PD1_ECM = Dict()
img = Array(Images.load(ECM_directory * filename * ".tif"))
for f in factors
    # get downsampled images
    img_small = imresize(img, ratio=1/f)
    img_re = imresize(img_small, size(img_small).*f)
    
    # save downsampled image
    plot(img_re, ticks = [], frame = :box, size = (500, 500))
    savefig(sample_CSV_directory * filename * "_factor_" * string(f) * ".png")
    
    # sample points
    img_mean_inv = 1- mean(Float64.(img_re))
    
    # compute number of points to sample
    n_sample = Int64(round(ECM_TDA.n_ECM_samples(img_mean_inv, low, high, min_sample, max_sample)))

    # sample points
    resampled, points, sampled, island_idx = sample_ECM_points(img_re, vicinity = 100, n_points = 5, n_samples = n_sample)

    # save sampled points to CSV
    df = DataFrame(resampled, [:x, :y])
    CSV.write(sample_CSV_directory * filename *  "_factor_" * string(f) * ".csv", df)
    
    # plot the results 
    scatter(resampled[:,1], resampled[:,2], yflip = :true, c = c_ECM, label = "", frame = :box, ticks = [], size = (500, 500))
    savefig(sample_CSV_directory * filename * "_factor_" * string(f) * "_points.png")
    
    # run PD
    df = CSV.read(sample_CSV_directory * filename * "_factor_" * string(f) * ".csv")
    PD0, PD1 = run_PH(df)
    # save
    save(sample_CSV_directory * filename * "_factor_" * string(f) * "PD.jld2", "PD0", PD0, "PD1", PD1)
    PD0_ECM[f] = PD0;
    PD1_ECM[f] = PD1;
end
```

```julia
# append the original PD 
# add the PD used in the default analysis
PD0_ECM["default"] = load("data/4000x4000_combined/ECM_PD/PD.jld2")["PD0_ECM"][filename]
PD1_ECM["default"] = load("data/4000x4000_combined/ECM_PD/PD.jld2")["PD1_ECM"][filename]

# convert array to Ripserer PD
PH0 = Dict(k => ECM_TDA.array_to_ripsererPD(v) for (k,v) in PD0_ECM if v != nothing)
PH1 = Dict(k => ECM_TDA.array_to_ripsererPD(v) for (k,v) in PD1_ECM if v != nothing)

# compute PI
PI0 = ECM_TDA.compute_PI(PH0);
PI1 = ECM_TDA.compute_PI(PH1);

save(sample_CSV_directory * "example_LTX001_Da232_idx1/" * filename * "_PI.jld2", 
    "PI0", PI0,
    "PI1", PI1)
```

```julia
max0 = get_PD0_max(PD0_ECM)
max1 = get_PD1_max(PD1_ECM)
```

## Compute average distance for 100 re-sampling from same resolution of image

```julia
# for reference, state the norm difference between different samples
# for each factor, create downsampled images & compute PD


PD0_ECM = Dict()
PD1_ECM = Dict()
img = Array(Images.load(ECM_directory * filename * ".tif"))
for f = 1:100
    
    # sample points
    img_mean_inv = 1- mean(Float64.(img))
    
    # compute number of points to sample
    n_sample = Int64(round(ECM_TDA.n_ECM_samples(img_mean_inv, low, high, min_sample, max_sample)))

    # sample points
    resampled, points, sampled, island_idx = sample_ECM_points(img_re, vicinity = 100, n_points = 5, n_samples = n_sample)

    # save sampled points to CSV
    df = DataFrame(resampled, [:x, :y])
    CSV.write("samples/" * filename *  "_sample_" * string(f) * ".csv", df)
    
    # run PD
    df = CSV.read("samples/"* filename * "_sample_" * string(f) * ".csv")
    PD0, PD1 = run_PH(df)

    PD0_ECM[f] = PD0;
    PD1_ECM[f] = PD1;
end
```

```julia
# convert array to Ripserer PD
PH0 = Dict(k => ECM_TDA.array_to_ripsererPD(v) for (k,v) in PD0_ECM if v != nothing)
PH1 = Dict(k => ECM_TDA.array_to_ripsererPD(v) for (k,v) in PD1_ECM if v != nothing)

# compute PI
PI0 = ECM_TDA.compute_PI(PH0);
PI1 = ECM_TDA.compute_PI(PH1);

save("samples/"* filename * "_PD_PI.jld2", 
    "PD0", PD0_ECM,
    "PD1", PD1_ECM,
    "PI0", PI0,
    "PI1", PI1)
```

```julia
# compute average distance between original (trial 1) and all others
mean_diff = [norm(PI1[1] - PI1[f]) for f = 1:100]
println("average distance for 100 sampling trials: ", mean(mean_diff))
```

# Run analysis for 100 randomly selected ECM images
1. Compute the distancees for each factor
2. Compute the average distance from different samples (as comparison)

```julia
# get all ROIs
dir = "data/4000x4000_combined/ECM_sampled/points_CSV/"
csv_files = [item for item in walkdir(dir)][1][3:end][1]

# sample 100
ROIs = sample(csv_files, 100; replace = false);

# parameters for sampling points from ECM image
low = 0.00063768
high = 0.4025769
min_sample = 100
max_sample = 5000

# directories
ECM_directory = "data/4000x4000_combined/subregion_ECM/"
sample_CSV_directory = "image_resolution_sensitivity/";

# factors to consider
factors = vcat([1], collect(4:4:128));
```

```julia
#save("image_resolution_sensitivity/ROIs.jld2", "ROI", ROIs)
```

```julia
# for each factor, create downsampled images & compute PD
for i = 1:100
    roi = ROIs[i]
    filename = roi[1:end-4]

    PD0_ECM = Dict()
    PD1_ECM = Dict()
    img = Array(Images.load(ECM_directory * filename * ".tif"))

    for f in factors
        # get downsampled images
        img_small = imresize(img, ratio=1/f)
        img_re = imresize(img_small, size(img_small).*f)

        # save downsampled image
        #plot(img_re, ticks = [], frame = :box, size = (500, 500))
        #savefig(sample_CSV_directory * filename * "_factor_" * string(f) * ".png")

        # sample points
        img_mean_inv = 1- mean(Float64.(img_re))

        # compute number of points to sample
        n_sample = Int64(round(ECM_TDA.n_ECM_samples(img_mean_inv, low, high, min_sample, max_sample)))
        if n_sample != 0
            # sample points
            resampled, points, sampled, island_idx = sample_ECM_points(img_re, vicinity = 100, n_points = 5, n_samples = n_sample)

            # save sampled points to CSV
            df = DataFrame(resampled, [:x, :y])
            CSV.write(sample_CSV_directory * "sampled_ECM/" * filename *  "_factor_" * string(f) * ".csv", df)

            # plot the results 
            #scatter(resampled[:,1], resampled[:,2], yflip = :true, c = c_ECM, label = "", frame = :box, ticks = [], size = (500, 500))
            #savefig(sample_CSV_directory * filename * "_factor_" * string(f) * "_points.png")

            # run PD
            #df = CSV.read(sample_CSV_directory *"sampled_ECM/"* filename * "_factor_" * string(f) * ".csv")
            PD0, PD1 = run_PH(df)

            PD0_ECM[f] = PD0;
            PD1_ECM[f] = PD1;
        else
            PD0_ECM[f] = nothing;
            PD1_ECM[f] = nothing;
        end
    end
    # save
    save(sample_CSV_directory * filename * "_PD.jld2", "PD0", PD0_ECM, "PD1", PD1_ECM)
end
```

Compute PI, PI distances and plot

```julia
for roi in ROIs

    PD =  load(sample_CSV_directory * roi[1:end-4] * "_PD.jld2")
    PD0 = PD["PD0"]
    PD1 = PD["PD1"]
    # add the PD used in the default analysis
    PD0["default"] = load("data/4000x4000_combined/ECM_PD/PD.jld2")["PD0_ECM"][roi[1:end-4]]
    PD1["default"] = load("data/4000x4000_combined/ECM_PD/PD.jld2")["PD1_ECM"][roi[1:end-4]]

    # convert array to Ripserer PD
    PH0 = Dict(k => ECM_TDA.array_to_ripsererPD(v) for (k,v) in PD0 if v != nothing)
    PH1 = Dict(k => ECM_TDA.array_to_ripsererPD(v) for (k,v) in PD1 if v != nothing)

    
    # add the PD used in the default analysis
    
    # compute PI
    if (PH0 != Dict{}()) & (PH1 != Dict{}())
        PI0 = ECM_TDA.compute_PI(PH0);
        PI1 = ECM_TDA.compute_PI(PH1);

        save(sample_CSV_directory * roi[1:end-4] * "_PI.jld2", 
            "PI0", PI0,
            "PI1", PI1)
    else
        save(sample_CSV_directory * roi[1:end-4] * "_PI.jld2", 
            "PI0", nothing,
            "PI1", nothing)
    end
end
```

compute PI1 differences at each factor

```julia
df = DataFrame([[],[],[],[]], [:factor,:roi, :diff, :SSIM])
for f in factors
    for roi in ROIs
        if roi  ["LTX097_Da113_idx2.csv"] # outlier
            PI1 = load(sample_CSV_directory * roi[1:end-4] * "_PI.jld2")["PI1"]
            if (PI1 != nothing)  
                if (f in keys(PI1))
                    ssim = assess_ssim(PI1["default"], PI1[f])
                    push!(df, [f,roi[1:end-4], norm(PI1["default"] - PI1[f]), ssim])
                end
            end
        end
    end
end
```

```julia
means = [mean(df[df.factor .== f, :diff]) for f in factors]
```

```julia
gr()
@df df boxplot(:factor, :diff, legend=false, 
                #ylims = [0, 0.005], 
                size = (900, 300),
                frame = :box,
                leftmargin = 5mm,
                bottommargin = 5mm,
                xlabel = "1 / scale factor",
                ylabel = "persistence image difference",
                c = "lightgrey")
plot!(factors, means, lw = 5, c = "#FF007F")
#savefig("image_resolution_sensitivity/sensitivity_plot.png")
```
